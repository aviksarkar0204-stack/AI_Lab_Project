```python
import os
import pandas as pd

class_path = []
file_path = []
for class_name in os.listdir(os.path.join("ISL_Dataset")):
    for file_name in os.listdir(os.path.join("ISL_Dataset", class_name)):
        file_path.append(os.path.join("ISL_Dataset", class_name, file_name))
        class_path.append(class_name)
        #print(class_name, file_path)
#df = pd.DataFrame(columns=["class_path", "file_path"])
#df['class_path'] = class_path
#df['file_path'] = file_path
df = pd.DataFrame(data={"class_path": class_path, "file_path": file_path})
```

```python
df.to_csv("image_paths.csv",index=False, sep=",")
```
